```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('CZ4046-Intelligent-Agents/assignment-one/results/part1_value_iteration_utility_estimates.csv', header=None)

# Extract state names and values
states = [f'State {i+1}' for i in range(len(df))]  # Generate state names
values = df.iloc[:, 1:].values.astype(float)

# Shift each row of values to start from 00.000
for i in range(len(values)):
    min_val = min(values[i])
    values[i] -= min_val

# Create a figure
plt.figure(figsize=(16, 8))  # Set the figsize here

# Iterate through each state and plot its values
for i, state in enumerate(states):
    # Extract the values for the current state
    state_values = values[i]
    
    # Plot the values
    plt.plot(state_values, label=state)

# Set x-axis limits to 0 and 700
plt.xlim(0, 700)

# Set logarithmic scale for y-axis
plt.yscale('log')

# Set y-axis limits to 1 and 100
plt.ylim(1, 100)

# Set labels and title
plt.xlabel('Number of Iterations')
plt.ylabel('Utility Value')
plt.title('Value Iteration')

# Add legend
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Show plot
plt.grid(True)
plt.show()
```
